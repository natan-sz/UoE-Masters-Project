## 3.2 Edge detection 

Eventually, our supervisor suggested the idea of analysing data collected by others in the lab. He suggested using data from Raman spectroscopy experiments from Phd students. One of the first things that was discussed was calculating the relative thickness of flakes of HfO2. This could be done by sampling one of the flakes and comparing the value to another flake. The more quantitative way of doing this would be to sum over all of the values, take an average (together with an uncertainty) and compare these values to the others. 

In order to achieve this, you would need to find the edges of the flakes to let the program automatically calculate these differences. 

From all of the algorithms I found, Canny's edge detection was the most useful. I looked into writing my own software to do this however it would be a waste of time when python has many libraries have it prewritten and as a prototype it should be satisfactory.

As the others had a few problems with getting this going, I wrote them a script using the opencv2 library to find the edge of the flakes. 

<pre><code>
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm
from scipy import ndimage as ndi

from skimage import feature

# Sources image, extracts green channel finds edges from parameters
src = cv2.imread("input.png", cv2.IMREAD_UNCHANGED)

im = src[:, :, 1]  # Green Channel
edges = np.array(255*feature.canny(im, sigma=9,
                                   low_threshold=0.01,
                                   high_threshold=0.50))

# Prints number of edge pixels
unique, counts = np.unique(edges, return_counts=True)
print(dict(zip(unique, counts)))

# Adds arrays together
result = np.add(im*-5, edges)

cmap = matplotlib.cm.bwr

plt.imshow(result, cmap=cmap)
plt.show()
</code></pre>

This script extracts the green channel from a given image (as this is the one of study). Loads it into a numpy array and runs the opencv canny method.

## 4.1 LAMMPS Input Script

This week I spent learning more about writing LAMMPS input scripts. My main source of this is the ARCHER LAMMPS tutorial series which goes through the use of LAMMPS and applying it for a HPC environment. This is the same tutorial that I used the simulation of Bulk Silicon.

ARCHER LAMMPS 2020 Tutorials:

* Part 1: https://www.youtube.com/watch?v=p3izaBP6gs0
* Part 2: https://www.youtube.com/watch?v=h2e_LYbls7g 

I used this supplemented with online forums to find out how to simulate more than one element. For this run I decided to stick with the simple cubic stucture (For real SiO2 it should be hexagonal {hpc}) for simplicity. 

This is the script I came up with.

<pre><code style="color:blue;">
# Title: Silicon Dioxide LAMMPS input script
# By: Natan Szczepaniak

# First input script which uses SIMPLE-NN generated potential
# to simulate Silicon Dioxide.


# 1) Setting up the simulation box

units		metal		# units
atom_style	atomic		# atoms are spheres
dimension	3		# 3 dimensions
boundary	p p p		# p - periodic boundaries 

lattice sc 2.65 		# sc - simple cubic
				# 0.6 - density 

region box block 0 10 0 10 0 10 # create a box of shape "block" (10x10)
create_box 2 box		# create box of atoms of 2 types

# Create Si atoms
create_atoms 1 region box

# Create O atoms
lattice sc 2.65 origin 0.3 0.3 0.3
create_atoms 2 region box

</code></pre>

First part of the script sets up the simulation box together with creating the atoms. 

<pre><code style="color:blue;">

# 2) Define interparticle ineractions (SIMPLE NN potential)

pair_style nn			# SIMPLE-NN potential file (.cpp)
pair_coeff * * /home/lammps_input/potential Si O # "* *" means every type interacts with every type with following coeffs

mass 1 28.0855			# Silicon mass 28.0855 grams/mol
mass 2 15.998			# Oxygen mass 15.998 grams/mol
</code></pre>

This is where the SIMPLE-NN potential is inputted and the mass of atoms is set. 

<pre><code style="color:blue;">

# 3) List of Neighbours (Neighbor lists are recalculated 
# each timestep.)

neighbor 2.0 bin		# Neighbour list 2.0 Angstrom

# neigh_modify delay 100 every 10
# neigh_modify delay 0 every 1 check yes page 500000 one 50000

</code></pre>

Neigbour lists make the simulation more efficient as it only considers atoms nearby.

<pre><code style="color:blue;">

# 4) Define Simulation parameters (Temperature, Momentum, etc.)

fix 3 all nvt temp 300.0 300.0 0.01
</code></pre>


This sets up the process in the simulation box. In this example, the box is simply kept at 300K


<pre><code style="color:blue;">
# 5) Final setup
#   - Define starting particle velocity
#   - Define timestep
#   - Define output system properties (temp, energy, etc.)
#   - Define simulation length
#
#velocity        all create 300 199085 mom no rot no
#
#timestep	0.005
#
#thermo_style 	custom step temp etotal pe ke press vol density # All of the outputs to log
#thermo 		10		# print at multiples of N
#
#run_style verlet
#
#run 200
#
# Output file

dump		2 all custom 10 positions.lammpstrj id type x y z vx vy vz
dump_modify	2 sort id

run 2000
</code></pre>

This step outputs all of the data into a file called positions.lammpstrj in the format given above and sorts the output by id. Alternatively you can also output this as a RDF plot or a Phonon spectrum. These outputs will be used to compare my simulations to the DFT calculations from the ab initio data for comparison. I will also visualise the structure to asses what is going on easier. 

In order to visualise and use my findings in OVITO, I need to specify the id, type, coordinates and velocities that will be output every timestep.
    
<code><pre>  
dump		2 all custom 10 positions.lammpstrj id type x y z vx vy vz
</pre></code>
    
Ensures the correct data is outputted however we can output whatever we would like Total energy, Potential Energy, etc. each timestep.
    
The form of this data "dump" as described in the LAMMPS documentation is:
    
<code><pre> 
dump		[ID] [Group-ID] [style]  [N] [filename] [args]
</pre></code>
   
As it states in the documentation:
    
> - ID = user-assigned name for the dump
>
> - group-ID = ID of the group of atoms to be dumped
>
> - style = atom or atom/gz or atom/zstd or *atom/mpiio or cfg or cfg/gz or cfg/zstd or cfg/mpiio or custom or custom/gz or custom/zstd or custom/mpiio or dcd or h5md or image or local or local/gz or local/zstd or molfile or movie or netcdf or netcdf/mpiio or vtk or xtc or xyz or xyz/gz or xyz/zstd or xyz/mpiio
>
> - N = dump every this many timesteps
>
> - filename = name of file to write dump info to
>
> - args = list of arguments for a particular style

We used a "custom" style as we want to specify our own terms for the output of the system.   
    
Additionally there is a difference between different outputs:
- thermo - output to console
- dump - output to file
- fix - set parameters for output to file
  
I kept the thermo ones in as it gives an output to the console whilst the simulation is running.

## 6.2 Bond Order Empirical Potentials (Tersoff)
    

    
As I had not much else to do after a milestone, I decided to learn more about Classical Potentials and how they can be used for analysis. In particular I looked at the Tersoff Potential.
    
Tersoff Potential is a type of a Empirical bond order interatomic potential. The advantage of this type of empirical potential that sets it apart from standard classical empirical potentials such as Lennard Jones is that it can model different bonding states of an atom.
    
This is dervied from the relation of bond strength and the bonding environment parameters. These include bond lengths, angles and the number of bonds.
    
The general form of the potential looks like this:
    
<!--\begin{split}
    E & = \frac{1}{2} \sum_i \sum_{j \neq i} V_{ij}
\end{split}--->
    
\begin{split}
V_{ij} & = f_C(r_{ij}) \left[ f_R(r_{ij}) + b_{ij} f_A(r_{ij}) \right] \end{split}
    
This potential consists of the repulsive two body repulsive term $f_R(r_{ij})$ and an attractive three-body term $f_A(r_{ij})$. All of this is enclosed by a cut off function $f_C(r_{ij})$ that sets the potential for each atom pair at $r>R+D$ where R+D is the cutoff distance. This is chosen to only include the first-neighbor shell.
    
The $f_A(r_{ij})$ three body term is multiplied by $b_{ij}$ is the bond order which depends on the environment of three particles. More on this later.
  
Here are the respective forms of each one of the functions:
    
\begin{split}
f_R(r) & =  A \exp (-\lambda_1 r) \\
f_A(r) & =  -B \exp (-\lambda_2 r) \\
f_C(r) & = \left\{ \begin{array} {r@{\quad:\quad}l}
  1 & r < R - D \\
  \frac{1}{2} - \frac{1}{2} \sin \left( \frac{\pi}{2} \frac{r-R}{D} \right) &
    R-D < r < R + D \\
  0 & r > R + D
  \end{array} \right. \\
\end{split}
    
$R,D,\lambda_1,$ and $\lambda_2$ are all parameters that need to be fed into the potential manually and are usually obtained by experimental/DFT calculations. By now, for majority of environments these parameters are mostly well established and are widely available for popular chemical systems such as Si-O and C-C.
    
As mentioned before, $f_A(r_{ij})$ is multiplied by the bond order term which relies on the three body interactions which are key to the bond strength modelled by the interatomic potential. This term can increase or decrease the attractive force which is modelling the bond. The magnitude of this term (hence strength of bond) is determined by the environemnt of the system for example it would be lowered when the number of neighbours is high and increase when number of bonds is relatively low. 

\begin{split}
b_{ij} & =  \left( 1 + \beta^n {\zeta_{ij}}^n \right)^{-\frac{1}{2n}} 
\end{split}
    
This relationship comes from the $\zeta_{ij}$ term.
    
\begin{split}
\zeta_{ij} & =  \sum_{k \neq i,j} f_C(r_{ik}) g(\theta_{ijk})
                 \exp \left[ {\lambda_3}^m (r_{ij} - r_{ik})^m \right] 
\end{split}

We see this from the exponent $\exp \left[ {\lambda_3}^m (r_{ij} - r_{ik})^m\right] $. The bigger the difference between $r_{ij}$ and $r_{ik}$ the bigger the $\zeta_{ij}$ term, the smaller the $b_{ij}$ therefore the weaker the bond.
    
Another important aspect of this potential is the angle of the bond (coordination of the system). The $\zeta_{ij}$ term includes a function $g(\theta_{ijk})$ where $\theta_{ijk}$ is the angle between the atoms $i,j$ and $k$.
     
\begin{split}
g(\theta) & =  \gamma_{ijk} \left( 1 + \frac{c^2}{d^2} -
                \frac{c^2}{\left[ d^2 + (\cos \theta - \cos \theta_0)^2\right]} \right)
\end{split}
    
The coordination of the three body system includes a lot of bond-specific terms, however the most important part is that the minimum for this function (where the bond is the strongest) is where $h=cos(\theta)$ as then the denominator goes to $d^2$ meaning $g(\theta)=1$.
    
###  Example (Carbon with 120$^o$ bonds)
    
As mentioned, the parameters for these systems is quite well established for systems such as Si or C. Here is a table of parameters for Tersoff Potentials I found [here](http://phycomp.technion.ac.il/~david/thesis/node24.html).
    
<img src="https://imgur.com/XqjHav6l.png" style="width:50%;height:auto;">
    
We see that $h=-0.598$ for carbon. At $120^o$ the denominator of $g(\theta)$ goes like this 
    
\begin{equation}
g(120^o)=1+\frac{c^2}{d^2}-\frac{c^2}{[d^2+(-0.598-cos120^o)^2]} \\
g(120^o)=1+\frac{c^2}{d^2}-\frac{c^2}{[d^2+(-0.598-(-0.5))^2]} \\
g(120^o)\approx 1+\frac{c^2}{d^2}-\frac{c^2}{d^2} \\
g(120^o)\approx 1
\end{equation}
    
This minimises $\zeta_{ij}$ and therefore maximises the $b_{ij}$ which corresponds to the bond strength (attractive force) making the bond strength maximal.

###  End of example
    
In the end, this is what the potential looks like. We can see that for a larger bond order term $b_{ij}$ the lower (negative) the potential meaning it is more attractive at a certain distance away from the particle.

## 8.2 DFT Summary Explanation

Since Joe Pitfield is not replying and I cannot move in with the project I would like to use this time to explain more about the theory behind DFT calculations for future reference. This will not be an exhaustive overview of DFT as it isnt the main focus of my project.
    
This was compiled using mostly a paper suggested by Joe and a talk "Fundamentals and applications of density functional theory - Astrid Marthinsen (Norwegian University) https://www.youtube.com/watch?v=SXvhDLCycxc". 
    
DFT in summary is the process of finding the lowest energy of a system from first principles calculations. The purpose of it is to replace the extremely demanding process of solving the Schrodinger equation for many body systems due to the great complexity. Schrodinger equation solutions are only accurate for Hydrogen atom as it is a 2 body system. For anything past 3 bodies, approximate methods are used.
    
DFT goes around this problem with a number of approximations. First one being the Born-Oppenheimer Approximation. This essentially means $m_{nuclei} >> m_{e}$. This essentially decouples the dynamics of the nucleus to the dynamics of the electrons. This means we can treat them seperately so the wavefunction looks like.
    
\begin{equation}
    \Psi({r_i}{R_i}) \rightarrow \Psi_N({R_i}) \ast \Psi_e({r_i})
\end{equation}
    
This is still extremely difficult to compute as it scales as 3N where N is the number of atoms. To solve this, we can utilise the Hohenberg-Kohn theorems.
    
### Hohenberg-Kohn Theorems
    
<ol>
    <li>Ground State Energy is a unique functional of the electron density.</li>
    <li>Electron density that minimises the energy functional gives the true ground state energy.</li>
</ol>
    
This means that we can use the electron density to get to the final lowest energy state.
    
\begin{equation}
    n(r) = \Psi\text{*}({r_i})  \Psi({r_i})
\end{equation}
    
This then takes the problem from 3N to 3 dimensions as only n(r) needs to be found. This is because the groundstate energy only depends on the electron density.
    
Through a minimisation algorithm we can obtain the grounds state of a system.

# 10.1  Revision of Behler-Parinello Method
    
As I am most likely going to be asked about this during the viva. I am going to revise the full method of Behler-Parinello as I have not read about it properly since last year considering I was learning all of the things mentioned previously.
    
This is generally the motto of this method:
  
> Regression model on the relationship between atomic configurations and total energies.
    
The reason for using this method is because it is nearly as accurate as DFT with the computational cost of Classical Simulations. Here is a qualitative graph showing this trend and where ML potentials fit in.

In part a) we see how firstly the cartesian coordinates are passed on to the symmetry functions to be used by the neural network. 
    
The main part of this entire method is making sure we get the correct symmetry functions to describe the system. In the iteration of the Behler-Parinello method I am using there are 3 symmetry functions (One radial(2 body), 2 Angular (3 body)) that are fitted to describe the environment around each atom in the simulation.
    
<img src="https://i.imgur.com/OZN6ZcN.png" style="width:30%:height:auto;">
    
Similarily to the Tersoff Potential, these are fitted with parameters to describe the system. This could be done manually however it would be counterintuitive to do it this way and it is simple enough to have code to do it for us. Fortunately SIMPLE-NN has a symmetry function fitting feature. The number of parameters to be optimised per symmetry function go like this radial G2(2 parameters) | angular G4(4 parameters) | G5(5 parameters) $\nu$ and $R_s$ are used for G2. The anuglar functions take in the previous two together with $\nu$, $\zeta$ and $\lambda$.
    
Once optimised, each type of symmetry function is saved as a vertex $G_i({R_i})$ in a python pickle file. Then the values are normalised between -1 and 1 and fed into the neural network. This gives the neural network the description of the local environment for each atom. (The symmetry functions are optimised using DFT data). There is a solution for fixing the sampling bias which is called GDF optimisation This prevents imperfections in the material in DFT calculations to transfer to the NN potential simulations. 
 
These normalised symmetry functions are then passed onto Tensorflow here test-train data is split and the neural network is trained using the Adam optimiser. There are other optimisers available however from what I have seen Adam is the standard as it is so ahead of the competition.
    
The simple-nn paper also mentiones that to speed up the computing time we can alter the loss function to only include the RMSE in Energy (exclude Force RMSE) however this apparently leads to bias with smaller datasets.
    
After the training is complete, the final form of the neural network is saved as "potential_saved_iteration" This contains the symmetry functions on top and the final state of the neural network in the end. This is then passed to LAMMPS in form of the custom "pair_style" command which is a template that takes in parameters (just as Tersoff just this time parameters are symmetry functions parameters and network weights).
    
This pair_style is a 900 line code of C++ code with little to no documentation so it is really a quest to find what works but from dissecting the code, I see that the code creates symmetry functions for each one of the particles at the time, inputs them into the neural network and obtains a Potential Energy Surface which is then used by LAMMPS to iterate a new step. The literature on this is quite limited and everything i have go off of is from the code so it may not be right.

mass * 12.0107

velocity all create 300.0 5432373 dist gaussian
